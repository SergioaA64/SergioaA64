## Exercícios de estatística Descritiva

* Faça a importação dos dados de clima de Brasília e Goiânia

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bsb = pd.read_csv(
    "Downloads/BRASILIAclima.csv",
    sep=";",
    parse_dates=["Data"],
    decimal=',',
    dayfirst=True
)
bsb.info()
```

```python
gyn = pd.read_csv(
    "Downloads/GOIANIAclima.csv",
    delimiter=";",
    parse_dates=["Data"],
    decimal=',',
    dayfirst=True
)
gyn.head()
```

* Calcule as medidas descritivas de média, desvio-padrão, coeficiente de variação e mediana para a temperatura instantanea em ambas as cidades.

```python
print(bsb["Temp. Ins. (C)"].mean())
print(bsb["Temp. Ins. (C)"].std())
print(bsb["Temp. Ins. (C)"].std()/bsb["Temp. Ins. (C)"].mean())
print(bsb["Temp. Ins. (C)"].median())
```

```python
print(gyn["Temp. Ins. (C)"].mean())
print(gyn["Temp. Ins. (C)"].std())
print(gyn["Temp. Ins. (C)"].std()/gyn["Temp. Ins. (C)"].mean())
print(gyn["Temp. Ins. (C)"].median())
```

* Calcule as médias das variáveis considerando um agrupamento diário para ambas as cidades

```python
media_diaria = bsb.groupby(bsb['Data']).mean().reset_index()
media_diaria["Cidade"]="BSB"
media_diaria.sort_values(by="Temp. Ins. (C)", ascending=False, inplace=True)
media_diaria.head()
```

```python
mdgyn = gyn.groupby(gyn['Data']).mean().reset_index()
mdgyn["Cidade"] = "GYN"
mdgyn.head()
```

```python
media_diaria =  media_diaria.append(mdgyn, ignore_index=True)
media_diaria.head()
```

* Apresente um gráfico de série temporal com a evolução das médias de temperatura diária ao longo do mes, usando cores diferentes para as duas cidades

```python
plt.figure(figsize=(14,5))
g = sns.lineplot(x = "Data", y =  "Temp. Ins. (C)", data=media_diaria, hue="Cidade")
g.set_title('Teste')
#plt.xticks(rotation=30)
#plt.show()
```

* Faça a categorização dos registros de umidade em 3 níveis conforme os seguintes intervalos:  muito seco (29.51, 40.028] , seco (40.028, 50.514] e razoável (50.514, 61.0]]

```python
def categoria(x):
    if x < 40.028:
        return 'muito seco'
    if x >= 40.028 and x < 50.514:
        return 'seco'
    if x >= 50.514:
        return 'razoável'
    
media_diaria['Categoria'] = media_diaria['Umi. Ins. (%)'].apply(categoria)
media_diaria
```

```python
pd.cut(media_diaria['Umi. Ins. (%)'], 3, labels=["muito seco", "seco", "razoável"])
```

* Apresente uma tabela de dupla entrada relacionando as cidades com as categorias de umidade do item anterior

```python
display(pd.crosstab(pd.cut(media_diaria['Umi. Ins. (%)'],3, labels=["muito seco", "seco", "razoável"]),media_diaria.Cidade))
pd.crosstab(media_diaria.Categoria, media_diaria.Cidade)
```
